```python
#Import Dependencies
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
#Set Column View
pd.set_option('display.max_columns',1000)
```

```python
#Import draft class
website_url = requests.get('https://watchstadium.com/news/up-to-date-list-of-college-basketball-players-who-declared-for-2019-nba-draft-04-01-2019/')
soup = BeautifulSoup(website_url.content,'lxml')
soup.prettify()
name_list = soup.find('div',{'class':"post--body"})
name_list
listdata = name_list.findAll('p')
data =[]
count = 0
for cell in listdata:
    count = count+1
    if (count >5):
        data.append(cell.get_text())
```

```python
#Put names and colleges in list and strip out college names
data2 = np.array(data)
```

```python
data2
```

```python
players = [items for item in data2 for items in item.split(',')]
# Deletes every second item in the list after the first item.  
##I.e. keeps all odd items in the list.
del players[1::2]
players
```

```python
# Retreive player name and get stats for players over their whole college career.
##Potential issue could be that it's for the whole college career, so players with 4 years of college could be hurt by freshman and sophmore years.
maindata = []
for name in players:
    realname = name
    name = name.lower()
    name = name.replace(" ", "-")
    #assumes player does not have the same name of a player previously drafted.
    website_url = requests.get('https://www.sports-reference.com/cbb/players/'+ name +'-1.html')
    soup = BeautifulSoup(website_url.content,'lxml')
    soup.prettify()
    My_table = soup.find('tr',{'id':'players_per_game.2019'})
    My_info = soup.find('div',{'id':'meta'})
    if My_table is not None:
        maindata = maindata + [realname]
        tabledata = My_table.findAll('td')
        infodata = My_info.findAll('p')
        data = []
        idata = []
        for cell in tabledata:
            data.append(cell.get_text())
        for thing in infodata:
            idata.append(thing.get_text(strip=True))
            #print(idata)
            idata = [x for x in idata if not 'RSCI'in x]
        maindata = maindata +data[0:26]+ idata[0:4]
```

```python
maindata
```

```python
len(players),len(maindata)
```

```python
maindata2 = np.array(maindata)
refined2 = np.reshape(maindata2, (-1, 31))
```

```python
My_table = soup.find('table',{'id':'players_per_game'})
tableheader = My_table.findAll('th')
colheaders = []
count = 0
for item in tableheader:
    if (count < 27):
        colheaders.append(item.get_text())
        count=count+1
colheaders = ['Name'] + colheaders + ['Position']+['Height_Weight']+['Hometown']+['High School']
print(colheaders)
```

```python
colheaders.remove('Season')
```

```python
college = pd.DataFrame(refined2, columns=colheaders)
```

```python
college
```

```python
college['Position']=college['Position'].str.replace("Position:","")
college['Hometown']=college['Hometown'].str.replace("Hometown:","")
college['High School']=college['High School'].str.replace("High School:","")
```

```python
boom = college['Height_Weight'].str.split(',',expand=True)
college['Height']=boom[0]
boom1 = boom[1].str.split('(',expand=True)
boom1[0] = boom1[0].str.replace("lb","")
college['Weight']= boom1[0]
```

```python
college.drop(columns={'Height_Weight'},inplace=True)
```
